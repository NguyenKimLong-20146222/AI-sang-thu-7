```python
import pandas as pd
url='/content/Advertising.csv'
data=pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['TV']), y = data['sales'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la:',model.coef_)
```

```python
model.score(X=pd.DataFrame(data['TV']), y = data['sales'])
```

```python
y_pred=model.predict(X=pd.DataFrame(data['TV']))
```

```python
data.plot(kind='scatter',x='TV', y = 'sales' , figsize=(9,9),color= 'black')
plt.plot(data['TV'],y_pred,color='blue')
```

```python
#Bc 2
model=linear_model.LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2]).T
model.fit(X, y=data['sales'])
print('Intercept ca m hnh:',model.intercept_) 
print('Coeffient ca m hnh:',model.coef_)
print(' chnh xc ca m hnh:',model.score(X,y=data['sales'])) 
```

```python
import numpy as np
range=np.arange(0,301,50)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```
